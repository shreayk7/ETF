```python
import requests
import pandas as pd

# Function to fetch ETF data
def fetch_etf_data(api_key, symbol):
    # Historical data
    url_time_series = f'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY_ADJUSTED&symbol={symbol}&apikey={api_key}'
    response = requests.get(url_time_series)
    data = response.json()

    # Profile data
    url_profile = f'https://www.alphavantage.co/query?function=ETF_PROFILE&symbol={symbol}&apikey={api_key2}'
    response2 = requests.get(url_profile)
    profile_data = response2.json()

    return data, profile_data

# Replace with your API key
api_key2="3Y076SBXD36WH4TZ"
api_key = "5OJV7FWOP5W8DAYR"
symbol = "QQQ"  # Example ETF symbol
time_series_data, profile_data = fetch_etf_data(api_key, symbol)

print(time_series_data)
print(profile_data)
```

```python
def preprocess_time_series(data):
    time_series = data.get("Monthly Adjusted Time Series", {})
    df = pd.DataFrame(time_series).T
    df = df.rename(columns={
        "1. open": "Open",
        "2. high": "High",
        "3. low": "Low",
        "4. close": "Close",
        "5. adjusted close": "Adj Close",
        "6. volume": "Volume",
        "7. dividend amount": "Dividend"
    })
    df.index = pd.to_datetime(df.index)
    df = df.sort_index()
    return df.astype(float)

time_series_df = preprocess_time_series(time_series_data)
print(time_series_df)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Scale data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series_df[["Adj Close"]])

# Create sequences
def create_sequences(data, seq_length):
    x, y = [], []
    for i in range(len(data) - seq_length):
        x.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(x), np.array(y)

seq_length = 30  # 30 days
X, y = create_sequences(scaled_data, seq_length)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define the model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(seq_length, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model.compile(optimizer="adam", loss="mean_squared_error")
model.summary()
```

```python
# Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data=(X_test, y_test))
```

```python
import plotly.graph_objs as go
import plotly.express as px

# Plot historical data
fig = go.Figure()
fig.add_trace(go.Scatter(x=time_series_df.index, y=time_series_df["Adj Close"],
                         mode='lines', name='Adjusted Close'))
fig.update_layout(title=" Adjusted Close Price Over Time",
                  xaxis_title="Date", yaxis_title="Price")
fig.show()
```

```python
# Make predictions
predictions = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predictions)

# Add predictions to visualization
fig.add_trace(go.Scatter(x=time_series_df.index[-len(predicted_prices):],
                         y=predicted_prices.flatten(),
                         mode='lines', name='Predicted Prices',
                         line=dict(color='red', dash='dash')))
fig.update_layout(title="Price Prediction",
                  xaxis_title="Date", yaxis_title="Price")
fig.show()
```

```python
import numpy as np
import pandas as pd
from datetime import timedelta

# Get the most recent sequence for prediction
last_sequence = X_test[-1]  # Shape (seq_length, 1)

# Number of steps to predict (e.g., 3 months = ~90 days for daily data)
future_steps = 90

# Create a placeholder for future predictions
future_predictions = []

# Recursive prediction
for _ in range(future_steps):
    # Predict the next price
    next_price = model.predict(last_sequence[np.newaxis, :, :])[0, 0]
    
    # Append the prediction
    future_predictions.append(next_price)
    
    # Update the sequence with the new prediction
    last_sequence = np.append(last_sequence[1:], [[next_price]], axis=0)

# Inverse transform the scaled predictions
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
```

```python
# Generate future dates
last_date = time_series_df.index[-1]
future_dates = [last_date + timedelta(days=i) for i in range(1, future_steps + 1)]

# Create a DataFrame for future predictions
future_df = pd.DataFrame({
    "Date": future_dates,
    "Predicted Price": future_predictions.flatten()
})
```

```python
import plotly.graph_objs as go

# Plot historical and predicted prices
fig = go.Figure()

# Historical data
fig.add_trace(go.Scatter(x=time_series_df.index, y=time_series_df["Adj Close"],
                         mode='lines', name='Adjusted Close'))

# Predicted data
fig.add_trace(go.Scatter(x=future_df["Date"], y=future_df["Predicted Price"],
                         mode='lines', name='Future Predictions',
                         line=dict(color='green', dash='dash')))

# Update layout
fig.update_layout(
    title="Price Prediction for Next 3 Months",
    xaxis_title="Date",
    yaxis_title="Price",
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

fig.show()
```
